```python
import pandas as pd
```

```python
df = pd.read_csv("libraries.csv")
df
```

```python
import geopy
```

```python
df['Zip'] = df['Zip'].astype('str') #note: just memorize that it's 'str' not 'string'
df['geo_address'] = df['Address'] + ", " + df['City'] + ", IN " + df['Zip'] + " United States"
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="library-coder")

from geopy.extra.rate_limiter import RateLimiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
df['location'] = df['geo_address'].apply(geocode)

df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
df
```

```python
df.to_csv("geocoded_libraries.csv")
```
